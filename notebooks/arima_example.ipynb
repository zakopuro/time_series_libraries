```python
import numpy as np
import pandas as pd
import pyflux as pf
import matplotlib.pyplot as plt
import datetime
```

```python
data = pd.read_csv('../data/sales_train_evaluation.csv.zip')
```

```python
all_sum = data.iloc[:,6:].sum().values
```

```python
plt.figure(figsize=(20,5))
plt.plot(all_sum[-56:])
plt.show()
```

```python
arima_model = pf.ARIMA(data=all_sum, ar=0, ma=4, family=pf.Normal())
```

```python
x = arima_model.fit("MLE")
```

```python
pred = arima_model.predict_is(h=28)['Series'].values
```

```python
plt.figure(figsize=(20,5))
plt.plot(np.arange(0,84),all_sum[-84:],label='True')
plt.plot(np.arange(56,84),pred,label='arima pred')
plt.legend()
plt.show()
```
