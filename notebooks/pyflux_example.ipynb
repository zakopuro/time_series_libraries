```python
import numpy as np
import pandas as pd
import pyflux as pf
import matplotlib.pyplot as plt
import datetime
```

```python
data = pd.read_csv('../data/sales_train_evaluation.csv.zip')
```

```python
all_sum = data.iloc[:,6:].sum().values
```

```python
plt.figure(figsize=(20,5))
plt.plot(all_sum[-56:])
plt.show()
```

```python
model = pf.ARIMA(data=all_sum, ar=8, ma=8, family=pf.Normal())
```

```python
x = model.fit("MLE")
```

```python
x.summary()
```

```python
model.plot_fit(figsize=(15, 10))
```

```python
model.plot_predict(h=28, past_values=56, figsize=(15,5))
```
